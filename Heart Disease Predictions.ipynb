```python
##Import relevant libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn as sns
```

```python
raw_data = pd.read_csv("C:/Users/Administrator/Downloads\Heart.csv")
```

```python
raw_data
```

```python
raw_data.describe()
```

```python
raw_data = raw_data.drop(columns = 'Unnamed: 0')
raw_data
```

```python
## One-hot encode categorical columns and ensure 0/1 output
raw_data = pd.get_dummies(raw_data, columns=['ChestPain', 'Thal'], drop_first=True, dtype=int)

##Encode target variable (AHD) as 0 and 1
raw_data['AHD'] = raw_data['AHD'].map({'Yes': 1, 'No': 0})
```

```python
raw_data
```

```python
raw_data.isnull().sum()
```

```python
data = raw_data.dropna()
data
```

```python
x = data.drop(columns = 'AHD')
x
```

```python
y = data['AHD']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 365)
```

```python
x_train
```

```python
x_test
```

```python
##Standardize the data

from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
x_train_scaled = scaler.fit_transform(x_train)
```

```python
x_test_scaled = scaler.transform(x_test)
```

```python
x_train_scaled
```

```python
x_test_scaled
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
log_reg = LogisticRegression(random_state = 42).fit(x_train_scaled, y_train)
```

```python
log_reg.predict(x_train_scaled)
```

```python
log_reg.score(x_train_scaled, y_train)
```

```python
log_reg.predict(x_test_scaled)
```

```python
log_reg.score(x_test_scaled, y_test)
```

```python
for c in [1.0, 0.5, 0.1, 0.01]:
    model = LogisticRegression(C=c, class_weight='balanced', max_iter=1000)
    model.fit(x_train_scaled, y_train)
    print(c, model.score(x_test_scaled, y_test))
```

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(log_reg, x_train_scaled, y_train, cv=5)
print("Average CV Accuracy:", scores.mean())
```

```python
log_reg = LogisticRegression(random_state = 42,
                             C = 0.01,
                             fit_intercept = True
                            ).fit(x_train_scaled, y_train)
```

```python
log_reg.score(x_train_scaled, y_train)
```

```python
log_reg.score(x_test_scaled, y_test)
```
